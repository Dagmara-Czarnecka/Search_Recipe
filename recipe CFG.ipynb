```python
import requests
import json

app_id = "b2dc5891"
app_key = "6fe030c54fb0b1173d0716519929adbc"

def search_recipes(ingredient, app_id, app_key, vegetarian=False, gluten_free=False):
    url = f"https://api.edamam.com/search?q={ingredient}&app_id={app_id}&app_key={app_key}"
    
    if vegetarian:
        url += "&health=vegetarian"
    
    if gluten_free:
        url += "&health=gluten-free"
    
    response = requests.get(url)
    data = response.json()

    if "hits" in data:
        recipes = data["hits"]
        recipes.sort(key=lambda x: x["recipe"]["calories"])
        return recipes
    else:
        return []

def display_recipes(recipes):
    if len(recipes) > 0:
        print(f"Found {len(recipes)} recipes:")
        for recipe in recipes:
            recipe_info = recipe["recipe"]
            print(f"\nTitle: {recipe_info['label']}")
            print("Ingredients:")
            for ingredient in recipe_info['ingredients']:
                print("- " + ingredient['text'])

            if 'url' in recipe_info and recipe_info['url']:
                print(f"Source URL: {recipe_info['url']}")
            else:
                print("Source URL not available for this recipe.")

            print(f"Calories: {recipe_info['calories']:.2f} kcal")
    else:
        print("No recipes found.")

def save_recipes_to_file(recipes, file_name):
    with open(file_name, 'w') as file:
        file.write(json.dumps(recipes))

def main():
    print("Welcome to Recipe Search!")
    ingredient = input("Enter an ingredient to search for: ")
    vegetarian = input("Do you want vegetarian recipes? (y/n): ").lower() == "y"
    gluten_free = input("Do you want gluten-free recipes? (y/n): ").lower() == "y"

    recipes = search_recipes(ingredient, app_id, app_key, vegetarian=vegetarian, gluten_free=gluten_free)
    display_recipes(recipes)

    save_option = input("Do you want to save the results to a file? (y/n): ")
    if save_option.lower() == "y":
        file_name = input("Enter a file name to save the results: ")
        save_recipes_to_file(recipes, file_name)
        print("Recipes saved to", file_name)

if __name__ == "__main__":
    main()
```
